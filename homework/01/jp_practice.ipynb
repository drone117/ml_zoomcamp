```python
u = np.array([2,4,5,6])
```

```python
v = np.array([1,0,0,2])
```

```python
np.dot(u, v) # dot product vector . vector
# dot product is used in algebra to find whether two vectors are close since
# a . b = |a||b|cos(f)
# if the lengths of both vectors are 1 then
# a . b = cos(f) <- the closer the dot product to 1 the more similar are the vectors in terms of direction
```

```python
assert u.shape[0] == v.shape[0]
res = 0
for i in range(u.shape[0]):
    res += u[i] * v[i]
```

```python
res
```

```python
u.dot(v)
```
